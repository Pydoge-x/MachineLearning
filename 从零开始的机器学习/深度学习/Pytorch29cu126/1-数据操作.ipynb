## 1.1.1 入门操作

1. 创建Tensor的方法:

方法|说明|参数
---|---|---
tensor|从Python列表或者Numpy数组创建Tensor|(data)
zeros|创建一个**全0**Tensor|(size)
ones|创建一个**全1**Tensor|(size)
empty|创建一个**未初始化**Tensor|(size)
rand|创建一个服从**均匀分布**的随机Tensor,值在[0, 1]|(size)
randn|创建一个服从**高斯分布**的随机Tensor,**均值为0**,**标准差为1**|(size)
arange|创建一个1维的序列Tensor，类似于Python的range|(start, end, step)
linspace|创建一个指定范围内等间隔的序列Tensor|(strat, end, steps)
eye|创建一个单位矩阵（只有对角线为1，其他全为0）|(size)
from_numpy|将NumPy数组转换为Tensor|(ndarray)

2. Tensor的属性:

属性|说明
---|---
shape|获取Tensor的形状
size()|获取Tensor的形状
dtype|获取Tensor的数据类型
device|查看Tensor所在的设备
dim()|获取Tensor的维度
requires_grad|是否启用梯度计算
numel()|获取Tensor中的元素总数
is_cuda|检查Tensor是否在GPU上
T|获取Tensor的转置
item()|获取Tensor的值
is_contiguous()|检查Tensor是否连续存储

```python
# 创建行向量
x = torch.arange(12)
x
```

```python
# 通过Tensor的shape属性查看形状
x.shape
```

```python
# 检查Tensor中的元素的总数
x.numel()
```

```python
# 将行向量修改为3行4列的矩阵
X = x.reshape(3, 4)
X
```

```python
# 创建全零Tensor
zeros_tensor = torch.zeros((2, 3, 4))
print(zeros_tensor)

# 创建全一Tensor
ones_tensor = torch.ones((2, 3, 4))
print(ones_tensor)
```

## 1.1.2 运算符

3. Tensor的操作：

操作|说明
---|---
+.-.*./.**|元素级别的加减乘除以及乘方运算
matmul|矩阵乘法
dot|**向量**点积(仅适用于一维Tensor)
sum|求和
mean|求均值
max|求最大值
min|求最小值
argmax(x, dim)|返回最大值的索引（指定维度）
softmax(x, dim)|计算softmax（指定维度）

```python
x = torch.tensor([1, 2, 4, 8])
y = torch.tensor([2, 2, 2, 2])

print(x+y)

print(x-y)

print(x*y)

print(x/y)  

print(x**y)
```

```python
# 求e的x次方
torch.exp(x)
```

4. Tensor的形状操作:

操作|说明
---|---
view(shape)|改变Tensor的形状(不改变数据)
reshape(shape)|类似于view但更灵活
t()|转置矩阵
unsqueeze(dim)|在指定维度添加一个维度
squeeze(dim)|去掉指定维度为1的维度
cat((x, y), dim)|连接多个Tensor

```python
X = torch.arange(12, dtype=torch.float32).reshape((3, 4))
Y = torch.tensor([[2.0, 1, 4, 3], [1, 2, 3, 4], [4, 3, 2, 1]])

# cat方法将指定维度连接成多个tensor。
torch.cat((X, Y), dim=0), torch.cat((X, Y), dim=1)
```

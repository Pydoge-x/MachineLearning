# 张量的操作

```python
import tensorflow as tf

a = tf.constant([[1,2],
                 [3,4]])
b = tf.constant([[5,6],
                [7,8]])

"""

基本运算：
    以下给出两种方法分别进行计算加减乘除
"""
# 方法1:
print("加法", (a + b).numpy())
print("减法", (a - b).numpy())
print("逐元素乘法", (a * b).numpy())
print("逐元素除法", (a / b).numpy())
```

```python
# 方法2：
print("加法",tf.add(a, b))
print("减法", tf.subtract(a, b))
print("逐元素乘法", tf.multiply(a, b))
print("逐元素除法", tf.divide(a, b))
```

```python
# 矩阵乘法
print(tf.matmul(a, b).numpy())
```

## 阶段总结
- 张量的数学运算有两种运算方式，如果计算两矩阵相乘，则使用`matmul`方法进行计算

```python
"""

矩阵操作

"""
c = tf.constant([[1,2,3],
                 [4,5,6]])
print("原始矩阵",c)
# 改变形状
print("转换为3*2矩阵", tf.reshape(c, [3,2]))

# 展平
print("将矩阵展平", tf.reshape(c, [-1]))

# 矩阵转置
print("转置后的矩阵", tf.transpose(c))
```

```python
# 增加/减少维度
print("增加维度", tf.expand_dims(c, axis=0)) # 在第0维增加维度
expand_tensor = tf.expand_dims(c, axis=0)
print(expand_tensor)
print("减少维度", tf.squeeze(expand_tensor, axis=0)) # 减少维度
```
